# AI Sentiment Analysis — Exploratory Analysis
Quick exploration of raw tweet data before the full pipeline.

```python
import sys
sys.path.insert(0, '../src')

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')
%matplotlib inline
```

## 1. Load Raw Data

```python
df_raw = pd.read_csv('../data/raw/raw_tweets.csv')
print(f'Shape: {df_raw.shape}')
df_raw.head()
```

## 2. Tweet Length Distribution

```python
df_raw['length'] = df_raw['content'].str.len()
df_raw['length'].hist(bins=40, color='steelblue', edgecolor='white')
plt.title('Tweet Character Length Distribution')
plt.xlabel('Characters')
plt.ylabel('Count')
plt.show()
```

## 3. Load Cleaned Data

```python
df_clean = pd.read_csv('../data/processed/clean_tweets.csv')
df_clean[['content', 'cleaned_text', 'processed_text', 'word_count']].head(10)
```

## 4. Sentiment Results

```python
df_labeled = pd.read_csv('../data/processed/labeled_tweets.csv')
df_labeled['sentiment'].value_counts().plot(kind='bar', color=['#4CAF50','#2196F3','#F44336'])
plt.title('Sentiment Distribution')
plt.xticks(rotation=0)
plt.show()
```

## 5. VADER Score Distribution

```python
df_labeled['vader_score'].hist(bins=50, color='coral', edgecolor='white')
plt.axvline(0.05, color='green', linestyle='--', label='positive threshold')
plt.axvline(-0.05, color='red', linestyle='--', label='negative threshold')
plt.title('VADER Compound Score Distribution')
plt.xlabel('Compound Score')
plt.legend()
plt.show()
```
